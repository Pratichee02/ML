```python
df=df.drop('Cabin',axis=1)
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean',missing_values=np.nan)

df['Age']=imp.fit_transform(df[['Age']])
df['Fare']=imp.fit_transform(df[['Fare']])
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Sex']=df['Sex'].replace('male',1)
df['Sex']=df['Sex'].replace('female',0)
```

```python
X=df.drop(['Survived','Embarked','Name','Ticket'],axis=1)
y=df.Survived
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print("Shape of training X ",Xtrain.shape)
print("Shape of training Y",ytrain.shape)
print("Shape of testing X ",Xtest.shape)
print("Shape of testing Y ",ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
Xtrain_tf=ss.fit_transform(Xtrain)
Xtest_tf=ss.fit_transform(Xtest)
```

```python
from sklearn.svm import SVC
svc=SVC(C=0.8,kernel='linear',gamma='auto')
svc.fit(Xtrain_tf,ytrain)
svc.fit(Xtest_tf,ytest)
score=svc.score(Xtrain_tf,ytrain)
scoret=svc.score(Xtest_tf,ytest)
```

```python
print(score)
print(scoret)
```

```python
y_pred=svc.predict(Xtest)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(ytest,y_pred)
cr=classification_report(ytest,y_pred)
```
