```python
X = data.drop(['Id', 'pH', 'free sulfur dioxide', 'chlorides', 'residual sugar', 'fixed acidity', 'quality'], axis=1)
```

```python
import seaborn as sns
sns.relplot(data=data, x="volatile acidity", y="alcohol", hue=y, palette="Set2")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_trainTF = scaler.fit_transform(X_train)
X_testTF = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

score = []
for i in range(1, 200):
    clf = KNeighborsClassifier(n_neighbors = i, metric="euclidean")
    clf.fit(X_trainTF, y_train)
    y_pred = clf.predict(X_testTF)
    score.append(accuracy_score(y_test, y_pred))

sns.relplot(data=score, kind="line")
```
